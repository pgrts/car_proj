```python
import requests
import pandas as pd
import numpy as np

def get_json(url):
    try:
        response = requests.get(url)
        response.raise_for_status()
        return response.json()
    except requests.exceptions.HTTPError as http_err:
        print(f"HTTP error occurred: {http_err}")
    except Exception as err:
        print(f"An error occurred: {err}")
    return {}

def vin_decode(vin):
    json_data = get_json(f'https://vpic.nhtsa.dot.gov/api/vehicles/decodevinvalues/{vin}?format=json')
    
    if json_data['Results']:
        df = pd.DataFrame(json_data['Results'])
    else: 
        df = pd.DataFrame(columns=cats+nums)
    return df
```

```python
def create_assumption(vin):
    
    df = vin_decode(vin)
    
    if df.empty:
        raise ValueError("Error: No vehicle information found for the provided VIN.")
        
    df = df.replace('','nan')

    
    required_fields = ['Make', 'Model', 'DisplacementCC', 'ModelYear']
    if df[required_fields].isnull().any().any():
        raise ValueError("Error: Make, Model, DisplacementCC, and ModelYear cannot be null.")
        
    # Check for VehicleType
    if 'VehicleType' not in df or df['VehicleType'].isnull().any():
        raise ValueError("Error: VehicleType cannot be null.")

    # Validate VehicleType
    allowed_vehicle_types = ['MULTIPURPOSE PASSENGER VEHICLE (MPV)', 'PASSENGER CAR', 'TRUCK']
    if df['VehicleType'].iloc[0] not in allowed_vehicle_types:
        raise ValueError("Error: VehicleType must be one of the following: MULTIPURPOSE PASSENGER VEHICLE (MPV), PASSENGER CAR, TRUCK.")

    # Check for BodyClass
    if 'BodyClass' not in df or df['BodyClass'].isnull().any():
        raise ValueError("Error: BodyClass cannot be null.")

    # Validate BodyClass
    disallowed_body_classes = ['Sport Utility Truck (SUT)', 'Roadster', 'Truck-Tractor', 'Truck', 'Limousine']
    if df['BodyClass'].iloc[0] in disallowed_body_classes:
        raise ValueError("Error: BodyClass cannot be one of the following: Sport Utility Truck (SUT), Roadster, Truck-Tractor, Truck, Limousine.")
        
    df['odometer'] = 100000
    df['paint_color'] = 'white'
    df['condition'] = 'good'
    df['state_income'] = 59802
    df['state'] = 'tx'
    df['region'] = 'dallas / fort worth'
    df['date_posted'] = -10

    return model_prep(df)
```

```python
def model_prep(df2):
    df2[cats] = df2[cats].astype(str)
    df2[nums] = df2[nums].astype('float64')
    return df2
```

```python
def find_similar_vehicles(row, df2, initial_threshold=0, increment=100, max_threshold=1000, n_veh=10):

    if isinstance(row, pd.DataFrame):
        row = model_prep(row)
        df1 = row.iloc[0]  # Ensure single row
    elif isinstance(row, pd.Series):
        df1 = row
    else:
        print('row is not series or dataframe')
        print(row)
        
    df2 = model_prep(df2)
    threshold = initial_threshold
    similar_vehicles = pd.DataFrame()

    while len(similar_vehicles) < n_veh and threshold <= max_threshold:
        similar_vehicles = df2[
            (df2['VehicleType'] == df1['VehicleType']) &
            (df2['DriveType'] == df1['DriveType']) &
            (df2['GVWR'] == df1['GVWR']) &
            (df2['BodyClass'] == df1['BodyClass']) &
            (df2['EngineCylinders'] == df1['EngineCylinders']) &
            (df2['ModelYear'] == df1['ModelYear']) &
            (abs(df2['DisplacementCC'] - df1['DisplacementCC']) < threshold) & 
            ((df2['Make'] + '_' + df2['Model']) != (df1['Make'] + '_' + df1['Model']))
        ]
        
        similar_vehicles = similar_vehicles.drop_duplicates(subset=['Make', 'Model'])

        if len(similar_vehicles) < n_veh:
            threshold += increment

    similar_vehicles['condition'] = df1['condition']
    similar_vehicles['state'] = df1['state']
    similar_vehicles['region'] = df1['region']
    similar_vehicles['state_income'] = df1['state_income']   
    
    
    return similar_vehicles.reset_index()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

def plot_comparison(row, df, model=cb72, cats=cats, nums=nums, odo_values=np.arange(50000, 300001, 10000)):
    # Prepare the row
    if isinstance(row, pd.DataFrame):
        row = model_prep(row)
        df1 = row.iloc[0]  # Ensure single row
    elif isinstance(row, pd.Series):
        df1 = row
    else:
        print('row is not series or dataframe')
        return None  # Return None if input is invalid
    
    label = create_label(df1)
    preds = create_odo_preds(df1, odo_values, model=model, cats=cats, nums=nums)

    # Initialize a DataFrame to hold the results
    results = pd.DataFrame({'Odometer': odo_values})
    results[label] = preds  # Use the primary vehicle label as the column name

    plt.figure(figsize=(12, 6))

    # Plot primary vehicle predictions
    plt.plot(odo_values, preds, label=label, color='red', linewidth=3)
    
    for idx, sim_row in df.iterrows():
        sim_label = create_label(sim_row)
        sim_preds = create_odo_preds(sim_row, odo_values, model=model, cats=cats, nums=nums)
        plt.plot(odo_values, sim_preds, label=sim_label, color='gray', linewidth=1)
        
        # Assign predictions to the results DataFrame for each similar vehicle
        results[sim_label] = sim_preds  # Directly assign predictions for the similar vehicle

        # Label predictions for similar vehicles every 50,000 miles
        for odo in range(0, 300001, 50000):
            if odo in odo_values:
                sim_pred_value = sim_preds[np.where(odo_values == odo)[0][0]]  # Get the corresponding prediction
                plt.text(odo, sim_pred_value, f'{int(round(sim_pred_value))}', fontsize=10, color='black')

    # Label predictions for the primary vehicle every 50,000 miles
    for odo in range(0, 300001, 50000):
        if odo in odo_values:
            pred_value = preds[np.where(odo_values == odo)[0][0]]  # Get the corresponding prediction
            plt.text(odo, pred_value, f'{int(round(pred_value))}', fontsize=10, color='black')

    # Add title and labels
    plt.title('Odometer Predictions vs. Odometer Values')
    plt.xlabel('Odometer Values (miles)')
    plt.ylabel('Predicted Values')

    # Show the legend
    plt.legend()

    # Display the plot
    plt.show()
    
    return results  # Return the DataFrame with results

# Example usage
results_df = plot_comparison(x1, sim_veh)
```

```python
import matplotlib.pyplot as plt

def plot_predictions_with_sim(x1, sim_veh, cb_model, cats, nums):
    
    odometer_values = np.arange(0, 300001, 10000)
    predictions_x1 = []
    predictions_sim = []

    for odo in odometer_values:
        x1['odometer'] = odo
        prediction_x1 = cb_model.predict(x1[cats + nums])
        predictions_x1.append(prediction_x1[0]) 

        sim_vehicle_predictions = []
        for idx in range(sim_veh.shape[0]):
            sim_veh_copy = sim_veh.copy()  # Create a copy to avoid SettingWithCopyWarning
            sim_veh_copy.iloc[idx, sim_veh_copy.columns.get_loc('odometer')] = odo  # Update odometer
            prediction_sim = cb_model.predict(sim_veh_copy[cats + nums])
            sim_vehicle_predictions.append(prediction_sim[0])  # Assuming the model returns an array-like structure
        
        predictions_sim.append(sim_vehicle_predictions)  # Store predictions for this odometer value

    # Set the figure size
    plt.figure(figsize=(12, 6))

    # Plot the predictions for x1
    plt.plot(odometer_values, predictions_x1, label="Prediction Line (x1)", color='red', linewidth=3)

    # Label predictions for x1 every 50,000 miles
    for i, odo in enumerate(odometer_values):
        if odo % 50000 == 0:
            plt.text(odo, predictions_x1[i], f'{int(predictions_x1[i])}', fontsize=10, color='black')

    # Colors for the simulated vehicles
    colors = ['blue', 'green', 'orange', 'purple', 'brown']  # Define a color palette for multiple vehicles

    # Plot predictions for each vehicle in sim_veh
    for idx, vehicle_predictions in enumerate(predictions_sim):
        plt.plot(odometer_values, vehicle_predictions, label=f"Prediction Line (Sim Vehicle {idx + 1})", 
                 color=colors[idx % len(colors)], linestyle='--', linewidth=1)

        # Label predictions for the current sim vehicle every 50,000 miles
        for i, odo in enumerate(odometer_values):
            if odo % 50000 == 0:
                plt.text(odo, vehicle_predictions[i], f'{int(vehicle_predictions[i])}', fontsize=10, 
                         color=colors[idx % len(colors)])

    # Add title and labels
    plt.title('Prediction vs Odometer Reading')
    plt.xlabel('Odometer (miles)')
    plt.ylabel('Price Prediction')

    # Show grid for better readability
    plt.grid(True)

    # Show the legend
    plt.legend()

    # Display the plot
    plt.show()

# Example usage:
plot_predictions_with_sim(x1, sim_veh, cb72, cats, nums)
```

```python
def clean_vin_output(df):

    df['price'] = df['price'].replace('$','').str.replace(',','')
    
    df[cats] = df[cats].astype(str)
    df[nums] = df[nums].astype('float64')
    df = df.dropna(subset=['Make','Model','ModelYear','VIN'])
    
    value_filter = (df.odometer > 50000) & (df.price > 5000) & (df.price < 90000) & (df.odometer < 300000) & (df.VehicleType.isin(['TRUCK', 'MULTIPURPOSE PASSENGER VEHICLE (MPV)', 'PASSENGER CAR'])) & (df['BodyClass'].isin(['Pickup',
       'Sport Utility Vehicle (SUV)/Multi-Purpose Vehicle (MPV)',
       'Crossover Utility Vehicle (CUV)', 'Sedan/Saloon',
       'Hatchback/Liftback/Notchback', 'Coupe', 'Convertible/Cabriolet',
       'Minivan', 'Wagon', 'Cargo Van', 'Van']))
       
    df2 = df[value_filter]
    df_bad = df[~value_filter]
    
    return df2, df_bad
```

```python
# Scrape Page 1 Every Week

from bs4 import BeautifulSoup
import time 

def df_from_link(link):
    response=requests.get(link)

    soup = BeautifulSoup(response.text, 'html.parser')

    # Finding all listing elements with class 'cl-static-search-result'
    listings = soup.find_all('li', class_='cl-static-search-result')

    # Preparing lists to store the data
    links = []
    prices = []
    locations = []

    # Loop through each listing and extract the required data
    for listing in listings:
        # Extracting link
        link_tag = listing.find('a', href=True)
        link = link_tag['href'] if link_tag else 'No link'

        # Extracting price
        price_tag = listing.find('div', class_='price')
        price = price_tag.text.strip() if price_tag else 'No price'

        # Extracting location
        location_tag = listing.find('div', class_='location')
        location = location_tag.text.strip() if location_tag else 'No location'

        # Append data to lists
        links.append(link)
        prices.append(price)
        locations.append(location)

    # Creating a DataFrame to store the extracted data
    dd = pd.DataFrame({
        'link': links,
        'price': prices,
        'location': locations
    })

    return dd

def scrape_regions(df):
    dfls = []

    for region_url in df.region_url.unique():
        link = region_url + '/search/cta?auto_title_status=1&bundleDuplicates=1&query=vin#search=1~gallery~0~0'
        tdf = df_from_link(link)
        tdf['region_url'] = region_url
        dfls.append(tdf)
        time.sleep(1)
        
    return dfls
```

```python
from datetime import datetime
import re

def link_parser(link):

    response=requests.get(link)
    soup = BeautifulSoup(response.text, 'html.parser')
    
    # Dictionary to hold the data
    data = {}

    # Define expected fields (labels) and initialize all as None
    fields = ['VIN:', 'condition:', 'drive:', 'fuel:', 'odometer:', 'paint color:', 'title status:', 'transmission:', 'type:']
    data = {field: None for field in fields}

    title_tag = soup.find('title')
        
    if title_tag:
        title = title_tag.text.strip()
        if title == 'blocked':
            return None
    else:
        title = None  # Just in case no title tag is found (shouldn't happen)
        
    time_tag = soup.find('time', class_='date timeago')
    if time_tag and 'datetime' in time_tag.attrs:
        posting_date_str = time_tag['datetime']
        try:
            # Use strptime for ISO format: "2024-10-18T17:06:07-0500"
            posting_date = datetime.strptime(posting_date_str, '%Y-%m-%dT%H:%M:%S%z')
        except ValueError:
            posting_date = None
        data['posting_date'] = posting_date
    else:
        data['posting_date'] = None
        
    geo_position_tag = soup.find('meta', attrs={'name': 'geo.position'})
    if geo_position_tag:
        geo_position = geo_position_tag.get('content', None)
        if geo_position:
            lat, long = geo_position.split(';')
            data['lat'] = lat
            data['long'] = long
    else:
        data['lat'] = None
        data['long'] = None

    geo_placename_tag = soup.find('meta', attrs={'name': 'geo.placename'})
    if geo_placename_tag:
        data['geo_placename'] = geo_placename_tag.get('content', None)
    else:
        data['geo_placename'] = None

    geo_region_tag = soup.find('meta', attrs={'name': 'geo.region'})
    if geo_region_tag:
        data['geo_region'] = geo_region_tag.get('content', None)
    else:
        data['geo_region'] = None
        
    posting_body_section = soup.find('section', id='postingbody')
    if posting_body_section:
        h2_tag = posting_body_section.find('h2')
        if h2_tag:
            # Clean text by removing HTML tags and special characters
            posting_body = h2_tag.get_text(strip=True)
            posting_body = re.sub(r'[^a-zA-Z0-9\s]', '', posting_body)  # Remove special characters
            data['postingbody'] = posting_body
        else:
            data['postingbody'] = None
    else:
        data['postingbody'] = None
        
        
    # Finding only the relevant divs with class "attrgroup"
    attr_groups = soup.find_all('div', class_='attrgroup')

    # Loop through each attrgroup and extract 'labl' and 'valu'
    for group in attr_groups:
        attrs = group.find_all('div', class_='attr')  # Find individual attributes in the group
        for attr in attrs:
            labl_tag = attr.find('span', class_='labl')
            valu_tag = attr.find('span', class_='valu')

            # Check if both labl and valu are present
            if labl_tag and valu_tag:
                labl = labl_tag.text.strip()  # Keep the original labl with the colon
                valu = valu_tag.text.strip()

                # Ensure we're only storing known fields
                if labl in data:
                    data[labl] = valu
                    
    data['title'] = title
    data['link'] = link
    
    ff = pd.DataFrame([data])
    return ff
```

```python
import statsmodels.api as sm
import patsy

# Define the formula for the regression model
formula = 'price ~ C(state) * C(BodyClass) * C(condition)'

# Use patsy to create the design matrices
y, X = patsy.dmatrices(formula, data=df2, return_type='dataframe')

# Fit the OLS regression model
model = sm.OLS(y, X).fit()

model.summary()
```
