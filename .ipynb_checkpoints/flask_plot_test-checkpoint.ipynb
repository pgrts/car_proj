```python
def get_json(url):
    try:
        response = requests.get(url)
        response.raise_for_status()
        return response.json()
    except requests.exceptions.HTTPError as http_err:
        print(f"HTTP error occurred: {http_err}")
    except Exception as err:
        print(f"An error occurred: {err}")
    return {}

def vin_decode(vin):
    json_data = get_json(f'https://vpic.nhtsa.dot.gov/api/vehicles/decodevinvalues/{vin}?format=json')
    df = pd.DataFrame(json_data['Results'])
    df.replace("", np.nan, inplace=True)
    df[new_cats +tryout_cats] = df[new_cats +tryout_cats].astype(str)
    df[[x for x in float_columns if x in df.columns]] = df[[x for x in float_columns if x in df.columns]].astype(float)
    clean_drive_type(df)
    return df


def clean_drive_type(row):
    
    if len(row == 1):
        drive_type = row['DriveType'][0]        
    else:
        print(row)
        print('not a dataframe with length 1')

    print(drive_type)
    if drive_type != 'nan':
        for d in ['RWD', '2WD', '4WD', 'FWD', 'AWD']:
            if d in drive_type: 
                row['DriveType'] =  d  
            
def find_similar_vehicles(df1, df2, initial_threshold=0, increment=100, max_threshold=1000):

    if isinstance(df1, pd.DataFrame):
        df1 = df1.iloc[0]  # Ensure single row

    threshold = initial_threshold
    similar_vehicles = pd.DataFrame()

    while len(similar_vehicles) < 5 and threshold <= max_threshold:
        similar_vehicles = df2[
            (df2['VehicleType'] == df1['VehicleType']) &
            (df2['DriveType'] == df1['DriveType']) &
            (df2['GVWR'] == df1['GVWR']) &
            (df2['BodyClass'] == df1['BodyClass']) &
            (df2['EngineCylinders'] == df1['EngineCylinders']) &
            (df2['ModelYear'] == df1['ModelYear']) &
            (abs(df2['displ'] - df1['displ']) < threshold) &
            (df2['Make'] != df1['Make']) &   # Ensure Make is different
            (df2['Model'] != df1['Model'])    # Ensure Model is different
        ]
        
        # Remove duplicates based on Make and Model to ensure uniqueness
        similar_vehicles = similar_vehicles.drop_duplicates(subset=['Make', 'Model'])

        if len(similar_vehicles) < 5:
            threshold += increment

    return similar_vehicles
```

```python
def generate_depreciation_curve(features, similar_vehicles, head_count=5):
    similar_vehicles = similar_vehicles.head(head_count)
    odometer_values = list(range(0, 240001, 6000))  # 0 to 240,000 miles with 6,000-mile steps
    predictions = []

    # Make predictions for different odometer values for the entered vehicle
    for odo in odometer_values:
        features['odometer'] = odo
        prediction = model.predict(pd.DataFrame(features))
        predictions.append(prediction[0])

    # Plot the depreciation curve
    plt.figure(figsize=(10, 6))
    plt.plot(odometer_values, predictions, label='Entered Vehicle', color='blue')

    # Plot depreciation curves for similar vehicles
    for i, vehicle in similar_vehicles.iterrows():
        vehicle_predictions = []
        for odo in odometer_values:
            vehicle_features = vehicle.to_dict()
            vehicle_features['odometer'] = odo
            vehicle_prediction = model.predict(pd.DataFrame([vehicle_features]))
            vehicle_predictions.append(vehicle_prediction[0])
        plt.plot(odometer_values, vehicle_predictions, label=f'Similar Vehicle {i+1}', linestyle='--')

    plt.title('Depreciation Curve')
    plt.xlabel('Odometer (Miles)')
    plt.ylabel('Price ($)')
    plt.legend()

    # Save plot to a string buffer
    buffer = io.BytesIO()
    plt.savefig(buffer, format='png')
    buffer.seek(0)
    img_base64 = base64.b64encode(buffer.getvalue()).decode()

    return img_base64
```

```python
def create_split_dc(split_method, ts=0.2, rs=194, stratify=True):
    if stratify:
        return {
            'tt' : train_test_split(X, y, test_size = ts, random_state = rs, stratify=y),
            'kf': StratifiedKFold(n_splits=5, shuffle=True, random_state=rs),
            'ss': StratifiedShuffleSplit(n_splits=5, test_size=ts, random_state=rs)
        }
    if not stratify:
        return {
            'tt' : train_test_split(X, y, test_size = ts, random_state = rs),
            'kf': KFold(n_splits=5, shuffle=True, random_state=rs),
            'ss': ShuffleSplit(n_splits=5, test_size=ts, random_state=rs)
        }        
    
def split_data(X, y, split_method, ts = 0.2, rs=194, stratify=True):

    split_dc = create_split_dc(split_method, ts=ts, rs=rs, stratify=stratify)    
    
    if stratify:        
        if split_method not in split_dc.keys():
            print(split_method + ' not in stratification dictionary')

        elif split_method == 'tt':
            X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = ts, random_state = rs, stratify=y)
            
        elif split_method in split_dc.keys():
            for train_index, test_index in split_dc[split_method].split(X, y):
                X_train, X_test = X.iloc[train_index], X.iloc[test_index]
                y_train, y_test = y.iloc[train_index], y.iloc[test_index]
        else:
            print(split_method + ' not allowed')
            return None        
            
    if not stratify:
    
        if split_method not in split_dc.keys():
            print(split_method + ' not in unstratified dictionary')

        elif split_method == 'tt':
            X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = ts, random_state = rs)
            
        elif split_method in split_dc.keys():
            for train_index, test_index in split_dc[split_method].split(X):
                X_train, X_test = X.iloc[train_index], X.iloc[test_index]
                y_train, y_test = y.iloc[train_index], y.iloc[test_index]
        else:
            print(split_method + ' not allowed')
            return None  
    return X_train, X_test, y_train, y_test
```
