```python
from fuzzywuzzy import fuzz
from fuzzywuzzy import process

def get_best_matches(ls1, ls2, threshold=90):
    best_matches = {}
    
    for item in ls1:
        # Find the best match for item from ls1 in ls2 with its score
        best_match, score = process.extractOne(item, ls2, scorer=fuzz.ratio)
        
        # Only include matches with a score greater than the threshold
        if score > threshold:
            best_matches[item] = best_match
    
    return best_matches

def fuzz_test3(ls1, ls2, thresh=50):
    
    for t1 in ls1:
        highest_score = 0
        best_match = None
        
        for t2 in l2:
            fuzz_score = fuzz.ratio(t1, mod)

            if fuzz_score > thresh and fuzz_score > highest_score:
                highest_score = fuzz_score
                best_match = t2
            
    return best_match, highest_score
```

```python
import pandas as pd
import numpy as np
import requests
import urllib.parse


#test = 'https://vpic.nhtsa.dot.gov/api/vehicles/decodevinvalues/1FTNW21F12ED23658?format=json'
def get_json(url):
    try:
        response = requests.get(url)
        response.raise_for_status()
        return response.json()
    except requests.exceptions.HTTPError as http_err:
        print(f"HTTP error occurred: {http_err}")
    except Exception as err:
        print(f"An error occurred: {err}")
    return {}
def vin_decode(vin):
    json_data = get_json(f'https://vpic.nhtsa.dot.gov/api/vehicles/decodevinvalues/{vin}?format=json')
    df = pd.DataFrame(json_data['Results'])
    df.replace("", np.nan, inplace=True)
    df.rename(columns={'DisplacementCC':'displ'})
    return df
```
