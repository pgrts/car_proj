```python
def get_json(url):
    try:
        response = requests.get(url)
        response.raise_for_status()
        return response.json()
    except requests.exceptions.HTTPError as http_err:
        print(f"HTTP error occurred: {http_err}")
    except Exception as err:
        print(f"An error occurred: {err}")
    return {}

#get_json('https://api.nhtsa.gov/products/vehicle/models?modelYear=2025&make=acura&issueType=r')
```

```python
# get all model for make/year : https://api.nhtsa.gov/products/vehicle/models?modelYear=2010&make=acura&issueType=r

def check_mdl_re(mk, yr):
    link = f'https://api.nhtsa.gov/products/vehicle/models?modelYear={yr}&make={mk}&issueType=r'
    json_data = get_json(link)
    
    if "results" in json_data:
        results = json_data["results"]
        rdf = pd.DataFrame(results, columns=["modelYear", "make", "model"])
        return rdf
    else:
        return None
```

```python
def add_product_columns(row):
    # Check if the row has a product and num_products == 1
    if row['num_products'] == 1 and isinstance(row['products'], list) and len(row['products']) > 0:
        # Extract the first dictionary (assuming one product per row when num_products == 1)
        product_info = row['products'][0]
        # Return the product info as a Series (with keys as column names)
        return pd.Series(product_info)
    else:
        # If no product or num_products != 1, return NaN for all columns
        return pd.Series({
            'type': np.nan,
            'productYear': np.nan,
            'productMake': np.nan,
            'productModel': np.nan,
            'manufacturer': np.nan
        })
```
