```python
def remove_outliers_by_rmse(df, threshold=1):
    """Iteratively remove outliers and track RMSE improvements, using a customizable threshold."""
    # Initialize variables
    df_outliers = pd.DataFrame(columns=df.columns)
    rmse_list = []  # List to track RMSE after each removal
    
    # Calculate initial RMSE for the entire dataset using your rmse function
    initial_rmse = rmse(df)
    previous_rmse = initial_rmse
    rmse_list.append(previous_rmse)
    
    while True:
        # Calculate the absolute errors
        df['error'] = abs(df['pred'] - df['price'])
        
        # Identify the row with the highest absolute error
        highest_error_row = df.loc[df['error'].idxmax()]
        
        # Remove the row with the highest error
        df_removed = df.drop(index=highest_error_row.name)
        
        # Calculate new RMSE after removing the highest error row
        new_rmse = rmse(df_removed)
        rmse_list.append(new_rmse)
        
        # Check if RMSE improved by more than the threshold, otherwise stop
        if abs(previous_rmse - new_rmse) < threshold:
            break
        
        # Update previous_rmse for the next iteration
        previous_rmse = new_rmse
        
        # Use pd.concat to add the highest error row to df_outliers
        df_outliers = pd.concat([df_outliers, highest_error_row.to_frame().T], ignore_index=True)
        
        # Update the dataframe for the next iteration
        df = df_removed
    
    # Return the good data and outliers
    df_good = df
    return df_good, df_outliers, rmse_list
```

```python
def validate_model_procedure(df, mult=1.5, add=1500, mult2=1.5, add2=1500, model=None):#, engine, datestr):
    df = model_prep(df)

    # STEP 2: LOAD OLDEST MODEL
    cb1 = CatBoostRegressor()
   
    if model:
        try:
            cb1.load_model(model)
        except:
            print('error ' + model + ' not found')
            return
    else:
        try:
            cb1.load_model(latest_cbm())
        except:
            print('error loading ' + latest_cbm())
            return
    
    y_pred = cb1.predict(df[cats+nums])
    df['pred'] =  np.round(y_pred).astype(int)

    length_1 = len(df)
    rmse_1 = rmse(df)

    df1, df2 = detect_down_payments(df,mult=mult, add=add)
    length_2 = len(df1)
    rmse_2 = rmse(df1)

    # error has increased, switch to op_mask with original_data
    if rmse_2 > rmse_1:
        df11 = df[~op_mask(df,mult=mult2, add=add2)]
        df12 = df[op_mask(df,mult=mult2, add=add2)]
        length_3 = len(df11)
        rmse_3 = rmse(df11)  
        
    # error as decreased, stay with df1, df2
    elif rmse_1 > rmse_2:
        df11 = df1[~op_mask(df1, mult=mult2, add=add2)]
        df12 = pd.concat([df2, df1[op_mask(df1,mult=mult2, add=add2)]])
        length_3 = len(df11)
        rmse_3 = rmse(df11)
    else:
        rmse_3 = rmse_1
        length_3 = length_1
    

    if rmse_3 > rmse_1:
        return df, pd.DataFrame(columns=df.columns), {length_1: rmse_1, length_2: rmse_2, length_3:rmse_3}
    elif rmse_3 < rmse_1:
        return df11, df12, {length_1: rmse_1, length_2: rmse_2, length_3:rmse_3}
    else:
        return df, pd.DataFrame(columns=df.columns), {length_1: rmse_1, length_2: rmse_2, length_3:rmse_3}

        
```

```python
from typing import List
from sqlalchemy import text

def create_table_from_schema(table_name: str, int_cols: List[str], float_cols: List[str], 
                             text_cols: List[str], dt_cols: List[str], engine):
    """
    Creates a PostgreSQL table with specified column types.

    Parameters:
    - table_name (str): The name of the table to create.
    - int_cols (List[str]): List of integer column names.
    - float_cols (List[str]): List of float (DOUBLE PRECISION) column names.
    - text_cols (List[str]): List of text column names.
    - dt_cols (List[str]): List of datetime column names (stored in ISO8601 format).
    - engine: SQLAlchemy engine for executing the statement.
    """
    # Define the schema creation string
    schema = f"CREATE TABLE IF NOT EXISTS {table_name} (\n"
    
    # Add primary key column
    schema += "    car_id SERIAL PRIMARY KEY,\n"
    
    # Add integer columns
    for col in int_cols:
        schema += f"    {col} INTEGER,\n"
    
    # Add float (double precision) columns
    for col in float_cols:
        schema += f"    {col} DOUBLE PRECISION,\n"
    
    # Add text columns
    for col in text_cols:
        schema += f"    {col} TEXT,\n"
    

    for col in dt_cols:
        schema += f"    {col} DATE,\n"
    
    # Remove the trailing comma and close parentheses
    schema = schema.rstrip(",\n") + "\n);"
    
    # Execute the schema creation statement
    with engine.connect() as conn:
        conn.execute(text(schema))
        conn.commit()
    
    print(f"Table '{table_name}' created successfully with the specified schema.")
```

```python
def prep_cd_sql(df, int_cols, float_cols, text_cols, dt_cols=['reference_date', 'date_scraped', 'posting_date']):

    #ii, ff, tt = remove_null_cols(null_cols, int_cols, float_cols, text_cols)
    df = df.replace({'None':np.nan, 'nan':np.nan})
    
    for col in int_cols:
        df[col] = pd.to_numeric(df[col], errors='coerce')
        df[col] = df[col].astype('Int64')
        df[col] = df[col].replace(-1, np.nan)

    for col in float_cols:
        df[col] = pd.to_numeric(df[col], errors='coerce')

    for col in ['reference_date', 'date_scraped']:
        df[col] = df[col].replace('None', np.nan)
        df[col] = pd.to_datetime(df[col])


    for col in text_cols:
        df[col] = df[col].replace('nan', np.nan)
        df[col] = df[col].replace('Not Applicable', np.nan)
        df[col] = df[col].replace('None', np.nan)
        
    df['posting_date'] = pd.to_datetime(df['posting_date'], format='ISO8601')
   # df['posting_date'] = df['posting_date'].dt.strftime('%Y_%m_%d')
    
    #df = df.replace('nan', np.nan)
    return df[int_cols+float_cols+text_cols+dt_cols]
```

```python
def calculate_rmse_for_models(dfs, models):
    results = []

    for df_name, df in dfs.items():
        length = len(df)
        
        for model_name, model in models.items():
            # Make predictions
            df['pred'] = np.round(model.predict(model_prep(df[cats + nums]))).astype(int)
            
            # Calculate RMSE
            rmse_value = rmse(df)
            
            # Append results to list
            results.append({
                'DataFrame': df_name,
                'Length': length,
                f'RMSE_{model_name}': rmse_value
            })

    # Convert results to DataFrame
    summary_df = pd.DataFrame(results)

    return summary_df
```

```python
def divert_price(valid_df, reject_df, big_rej_df, match_col='vin'):

    match_len = len(valid_df) + len(reject_df)
    
    # Find rows in valid_df that also exist in big_rej_df based on both match_col and price
    matching_rows = valid_df.merge(big_rej_df[[match_col, 'price']].drop_duplicates(subset=match_col), on=[match_col, 'price'], how='inner')
    
    # Add matching rows to reject_df
    reject_df = pd.concat([reject_df, matching_rows], ignore_index=True)
    
    # Filter valid_df to remove rows that match both match_col and price in matching_rows
    valid_df = valid_df.merge(matching_rows[[match_col, 'price']], on=[match_col, 'price'], how='left', indicator=True)
    valid_df = valid_df[valid_df['_merge'] == 'left_only'].drop(columns=['_merge'])

    test_len = len(valid_df) + len(reject_df)
    if match_len != test_len:
        print('ERROR ERROR' + f'transformed dataframe length: {test_len} does not match initial dataframe length: {match_len}')
        return None, None
    else:
        return valid_df, reject_df
```

```python
def df_from_link(link):
    response=requests.get(link)

    soup = BeautifulSoup(response.text, 'html.parser')

    # Finding all listing elements with class 'cl-static-search-result'
    listings = soup.find_all('li', class_='cl-static-search-result')

    # Preparing lists to store the data
    links = []
    prices = []
    locations = []

    # Loop through each listing and extract the required data
    for listing in listings:
        # Extracting link
        link_tag = listing.find('a', href=True)
        link = link_tag['href'] if link_tag else 'No link'

        # Extracting price
        price_tag = listing.find('div', class_='price')
        price = price_tag.text.strip() if price_tag else 'No price'

        # Extracting location
        location_tag = listing.find('div', class_='location')
        location = location_tag.text.strip() if location_tag else 'No location'

        # Append data to lists
        links.append(link)
        prices.append(price)
        locations.append(location)

    # Creating a DataFrame to store the extracted data
    dd = pd.DataFrame({
        'link': links,
        'price': prices,
        'location': locations
    })

    return dd
```

```python
def link_parser(link):
    response=requests.get(link)
    soup = BeautifulSoup(response.text, 'html.parser')

    # Dictionary to hold the data
    data = {}

    # Define expected fields (labels) and initialize all as None
    fields = ['VIN:', 'condition:', 'drive:', 'fuel:', 'odometer:', 'paint color:', 'title status:', 'transmission:', 'type:']
    data = {field: None for field in fields}

    title_tag = soup.find('title')
        
    if title_tag:
        title = title_tag.text.strip()
        if title == 'blocked':
            return None
    else:
        title = None  # Just in case no title tag is found (shouldn't happen)
        
    time_tag = soup.find('time', class_='date timeago')
    if time_tag and 'datetime' in time_tag.attrs:
        posting_date_str = time_tag['datetime']
        try:
            # Use strptime for ISO format: "2024-10-18T17:06:07-0500"
            posting_date = datetime.strptime(posting_date_str, '%Y-%m-%dT%H:%M:%S%z')
        except ValueError:
            posting_date = None
        data['posting_date'] = posting_date
    else:
        data['posting_date'] = None
        
    geo_position_tag = soup.find('meta', attrs={'name': 'geo.position'})
    if geo_position_tag:
        geo_position = geo_position_tag.get('content', None)
        if geo_position:
            lat, long = geo_position.split(';')
            data['lat'] = lat
            data['long'] = long
    else:
        data['lat'] = None
        data['long'] = None

    geo_placename_tag = soup.find('meta', attrs={'name': 'geo.placename'})
    if geo_placename_tag:
        data['geo_placename'] = geo_placename_tag.get('content', None)
    else:
        data['geo_placename'] = None

    geo_region_tag = soup.find('meta', attrs={'name': 'geo.region'})
    if geo_region_tag:
        data['geo_region'] = geo_region_tag.get('content', None)
    else:
        data['geo_region'] = None

    # Finding only the relevant divs with class "attrgroup"
    attr_groups = soup.find_all('div', class_='attrgroup')

    # Loop through each attrgroup and extract 'labl' and 'valu'
    for group in attr_groups:
        attrs = group.find_all('div', class_='attr')  # Find individual attributes in the group
        for attr in attrs:
            labl_tag = attr.find('span', class_='labl')
            valu_tag = attr.find('span', class_='valu')

            # Check if both labl and valu are present
            if labl_tag and valu_tag:
                labl = labl_tag.text.strip()  # Keep the original labl with the colon
                valu = valu_tag.text.strip()

                # Ensure we're only storing known fields
                if labl in data:
                    data[labl] = valu
                    
    data['title'] = title
    data['link'] = link
    
    ff = pd.DataFrame([data])
    return ff

def clean_listing_output(df2, datestr):
    # Column renaming dictionary
    column_rename_dc = {
        'VIN:': 'vin', 'condition:': 'condition', 'drive:': 'drive', 'fuel:': 'fuel',
        'paint color:': 'paint_color', 'type:': 'type', 'transmission:': 'transmission',
        'title status:': 'title_status', 'odometer:': 'odometer'
    }

    # Rename and clean DataFrame
    df = df2.rename(columns=column_rename_dc).reset_index(drop=True)
    df = df.replace('', 'nan')

    if 'odometer' in df and df['odometer'].dtype == 'object':
        # Replace commas and handle non-numeric values
        df['odometer'] = df['odometer'].str.replace(',', '', regex=False)  # Remove commas
        
        # Convert to numeric, forcing errors to NaN (useful for invalid or missing values)
        df['odometer'] = pd.to_numeric(df['odometer'], errors='coerce')

        # Optionally, you can fill NaN values with a default value (e.g., -1 or 0) if needed
        df['odometer'] = df['odometer'].fillna(-1).astype(int)  # Replace NaNs with -1 and convert to int

    # Truncate VIN to 16 characters if longer, else mark for rejection if less than 16
    df['vin'] = df['vin'].str[:16]
    df['date_scraped'] = datestr
    # Filter listings with valid VIN and odometer
    valid_df = df[(df['vin'].notnull()) & 
                  (df['vin'].str.len() == 16) & 
                  (df['odometer'].between(10000, 400000))]
    reject_df = df[~df.index.isin(valid_df.index)]  # Entries that don't meet criteria

    return valid_df, reject_df
```

```python
def divert_price(valid_df, reject_df, big_rej_df, match_col='vin'):
    # Find matching rows in both valid_df and big_rej_df by 'match_col' and 'price'
    matching_rows = valid_df.merge(big_rej_df[[match_col, 'price']], on=[match_col, 'price'], how='inner')
    
    # Add matching rows to reject_df
    reject_df = pd.concat([reject_df, matching_rows], ignore_index=True)
    
    # Remove rows in valid_df that have matching 'vin' and 'price' in matching_rows
    valid_df = valid_df[~valid_df[match_col].isin(matching_rows[match_col]) | 
                        ~valid_df['price'].isin(matching_rows['price'])]
    
    return valid_df, reject_df
```

```python
def check_links_and_scrape_price(df, link_column='link', price_column='price'):
    # Create a new DataFrame to store the rows with new and old prices
    new_df = pd.DataFrame(columns=['modelyear', 'make', 'model', 'old_price', 'new_price', 'link'])

    for index, row in df.iterrows():
        # Check if 'link_column' exists and is not empty
        if pd.notna(row[link_column]):
            link = row[link_column]

            try:
                # Request the link
                response = requests.get(link)
                
                # If the link is valid (status code 200)
                if response.status_code == 200:
                    # Parse the page with BeautifulSoup
                    soup = BeautifulSoup(response.content, 'html.parser')

                    # Scrape the price (this may vary based on the page structure)
                    # Assuming the price is contained in an element with a class 'price'
                    price_tag = soup.find(class_='price')  # Adjust this based on the actual page structure

                    if price_tag:
                        new_price = float(price_tag.get_text().replace('$', '').replace(',', '').strip())
                        
                        # Check if the new price is different from the existing price in the row
                        if new_price != row[price_column]:
                            # Create a new row with old and new prices
                            new_row = row.copy()
                            new_row['old_price'] = row[price_column]
                            new_row['new_price'] = new_price
                            new_df = new_df.append(new_row, ignore_index=True)
            except requests.exceptions.RequestException as e:
                # Handle possible exceptions, like network issues, invalid URLs, etc.
                print(f"Error requesting {link}: {e}")

    return new_df

# Example usage:
# Assuming 'df' is your DataFrame and it contains the columns 'link' and 'price'
new_price_df = check_links_and_scrape_price(df)
print(new_price_df)
```
