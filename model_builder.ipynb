```python
def create_split_dc(split_method, ts=0.2, rs=194, stratify=True):
    if stratify:
        return {
            'tt' : train_test_split(X, y, test_size = ts, random_state = rs, stratify=y),
            'kf': StratifiedKFold(n_splits=5, shuffle=True, random_state=rs),
            'ss': StratifiedShuffleSplit(n_splits=5, test_size=ts, random_state=rs)
        }
    if not stratify:
        return {
            'tt' : train_test_split(X, y, test_size = ts, random_state = rs),
            'kf': KFold(n_splits=5, shuffle=True, random_state=rs),
            'ss': ShuffleSplit(n_splits=5, test_size=ts, random_state=rs)
        }        
```

```python
def split_data(X, y, split_method, ts = 0.2, rs=194, stratify=True):

    split_dc = create_split_dc(split_method, ts=ts, rs=rs, stratify=stratify)    
    
    if stratify:        
        if split_method not in split_dc.keys():
            print(split_method + ' not in stratification dictionary')

        elif split_method == 'tt':
            X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = ts, random_state = rs, stratify=y)
            
        elif split_method in split_dc.keys():
            for train_index, test_index in split_dc[split_method].split(X, y):
                X_train, X_test = X.iloc[train_index], X.iloc[test_index]
                y_train, y_test = y.iloc[train_index], y.iloc[test_index]
        else:
            print(split_method + ' not allowed')
            return None        
            
    if not stratify:
    
        if split_method not in split_dc.keys():
            print(split_method + ' not in unstratified dictionary')

        elif split_method == 'tt':
            X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = ts, random_state = rs)
            
        elif split_method in split_dc.keys():
            for train_index, test_index in split_dc[split_method].split(X):
                X_train, X_test = X.iloc[train_index], X.iloc[test_index]
                y_train, y_test = y.iloc[train_index], y.iloc[test_index]
        else:
            print(split_method + ' not allowed')
            return None  
    return X_train, X_test, y_train, y_test
```
